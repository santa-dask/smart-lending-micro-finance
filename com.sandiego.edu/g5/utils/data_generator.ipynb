```python
!pip install Faker
import random
import pandas as pd
from faker import Faker

fake = Faker('en_IN')  # Indian English locale

# Function to generate random data for each variable
def generate_data():
    professions = ['Boutique Owner','Carpenter', 'Accountant', 'Digital_Marketer', 'Massage_Therapists', 'Electricians', 'Caterering_Service_Owner', 'Content_Creator', "Food_Chain_Owner", "Digital_Marketing"] # added list of professions as per Santo's suggestion
    borrower_data = {
        'borrower_name': fake.name(),
        'borrower_uid': fake.uuid4(),
        'borrower_age': random.randint(18, 65),
        'borrower_income': round(random.uniform(100000, 2000000), 2),
        #'borrower_profession': fake.job(),
        'borrower_profession': random.choice(professions), # select a random profession from list
        'borrower_address': fake.address(),
        'borrower_pan': ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=10)),
        'borrower_pincode': fake.postcode(),

        'business_name': fake.company(),
        'business_address': fake.address(),
        'business_uid': fake.uuid4(),
        'business_TAN': ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=10)),
        'business_pan': ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=10)),
        'business_annual_income': round(random.uniform(500000, 50000000), 2),
        'business_annual_revenue': round(random.uniform(1000000, 100000000), 2),
        'business_annual_expense': round(random.uniform(100000, 50000000), 2),
        'business_age': random.randint(1, 50),
        'business_sector': fake.bs(),
        'business_no_of_employees': random.randint(1, 1000),
        'business_revenue_1year': round(random.uniform(1000000, 100000000), 2),
        'business_revenue_2year': round(random.uniform(1000000, 100000000), 2),
        'business_revenue_3year': round(random.uniform(1000000, 100000000), 2),
        'business_revenue_4year': round(random.uniform(1000000, 100000000), 2),

        'loan_amount_asked': round(random.uniform(100000, 10000000), 2),
        'business_credit_score': random.randint(300, 900),
        'loan_amount_offered': round(random.uniform(100000, 10000000), 2),
        'loan_interest_rate': round(random.uniform(5, 20), 2),
        'loan_duration_months': random.randint(12, 360),
        'loan_monthly_payment': round(random.uniform(1000, 100000), 2),
        'business_default': random.choice([True, False]),
        'business_previous_loan': random.choice([True, False]),
        'business_previous_loan_paid': random.choice([True, False]),
        'borrower_business_address_match': random.choice([True, False]),
        'final_credit_score': random.randint(300, 900),
        'final_loan_verdict': random.choice(['Approved', 'Rejected']),
        'current_loan_request_amount': round(random.uniform(100000, 10000000), 2),
        'duration': random.randint(1, 30),

        # Social media and reviews
        'linkedin_profile': fake.url(),
        'facebook_profile': fake.url(),
        'business_reviews': random.randint(1, 5),
        'business_review_count': random.randint(0, 1000),
        'business_social_media_mentions': random.randint(0, 500)
    }

    return borrower_data

# Generate sample data for multiple entries
data = [generate_data() for _ in range(100000)]  # Generate 100 entries

# Convert to DataFrame
df = pd.DataFrame(data)

# Export to CSV
df.to_csv('loan_data_with_social_media.csv', index=False)

print("Data exported to loan_data_with_social_media.csv")
```
